## i. Bernoulli Distribution
A trail having two mutually exclusive and exhaustive outcomes is called Bernoulli trial. These two outcomes are generally known as success and failure. A random variable X that takes values 0 and 1 with probability of failure q and probability of success p respectively is called Bernoulli Variate and the distribution, it followed is called Bernoulli distribution.

A random variable X is saaid to have Bernoulli distribution if its probability mass function is given by:
$$P(x) = p^{x}*q^{1-x}, x = 0, 1$$
Here, p is parameter of the distribution. A random variable following Bernoulli distribution is denoted by b(1,p).

### Moments about origin of Bernoulli Distribution
When $X\sim{b}(1,p)$  then pmf of Bernoulli distribution is $p(x)=p^{x}q^{1-x}, x=0,1 $ then moments about origin is $$\mu_{r}' = E(x^{r}) = \sum_{x=0}^{1}x^{r}p(x)=\sum_{x=0}^{1}{x^{r}p^{x}q^{1-x}}, r = 1,2,3,4$$
When r = 1
$$\mu_{1}' = \sum_{x = 0}^{1}x^{1}p^{x}q^{1-x}= 0 + p = p$$
When r = 2
$$\mu_{2}' = \sum_{x = 0}^{1}x^{2}p^{x}q^{1-x}= 0 + p = p$$
When r = 3
$$\mu_{3}' = \sum_{x=0}^{1}x^{3}p^{x}q^{1-x}= 0 + p = p$$
When r = 4
$$\mu_{4}' = \sum_{x=0}^{1}x^{4}p^{x}q^{1-x}= 0 + p = p$$

```python
import numpy as np

import matplotlib.pyplot as plt

def bernoulli_distribution(p):
    q = 1 - p
    x = np.array([0, 1])
    y = np.array([q, p])
    return x, y

# Example usage
p = 1/3
x, y = bernoulli_distribution(p)

plt.bar(x, y, tick_label=['0 (Failure)', '1 (Success)'])
plt.xlabel('Outcome')
plt.ylabel('Probability')
plt.title('Bernoulli Distribution (p={})'.format(p))
plt.show()
```

```python
# Calculate moments about the origin for Bernoulli distribution
def moments_about_origin(x, y, k):
    return np.sum(x**k * y)

# Example usage
moments = [moments_about_origin(x, y, k) for k in range(1, 5)]

# Plot the moments
plt.bar(range(1, 5), moments, tick_label=['1st', '2nd', '3rd', '4th'])
plt.xlabel('Moment')
plt.ylabel('Value')
plt.title('Moments about Origin for Bernoulli Distribution (p={})'.format(p))
plt.show()
```

## Moments about mean(central moments) of Bernoulli distribution
The moments about mean of Bernoulli distribution can be obtained by using relationship between central moments and raw moments
$$\mu_{1} = \mu_{1}' - \mu_{1}' = 0\\
\mu_{2} = \mu_{1}' - \mu_{1}^{2} = p - p^{2} = p(1-p) = pq\\
\mu_{3} = \mu_{3}' - 3\mu_2'\mu_{1}' + 2\mu_{1}'^{3} = p - 3p^{2} + 2p^{3}\\
\mu_{4} = \mu_{4}' - 4\mu_{3}'\mu_{1} + 6\mu_{2}'\mu_{1}^{2} - 3\mu_{1}'^{4} = p - 4 pp + 6pp^{2}-3p^{4} = p - 4p^{2}+ 6p^{3} - 3p^{4}$$
 

```python
# This program finds the moments about mean(central moments of Bernoulli distribution) and plot them.
import numpy as np
import matplotlib.pyplot as plt
p = 0.6
mu1 = 0
mu2 = p*(1-p)
mu3 = p - 3*p**2 + 2*p**3
mu4 = p - 4*p**2 + 6 * p**3 - 3*p**4
print('The moments about mean(central moments) of Bernoulli distribution are : ')
print(f"$$\mu_{1}$$ = {mu1}")
print(f"$$\mu_{2}$$ = {mu2}" )
print(f"$$\mu_{3}$$ = {mu3}")
print(f"$$\mu_{4}$$ = {mu4}")
# Printing the mean and variance of Bernoulli distribution
```

## Moment Generating Function of Bernoulli Distribution

The moment generating function (MGF) of a random variable $X$ is defined as $M_X(t) = E(e^{tX})$. For a Bernoulli random variable $X$ with parameter $p$, the MGF is given by:

$$M_X(t) = E(e^{tX}) = \sum_{x=0}^{1} e^{tx} P(X=x)$$

Since $X$ can take values 0 and 1 with probabilities $q = 1 - p$ and $p$ respectively, we have:

$$M_X(t) = e^{t \cdot 0} \cdot q + e^{t \cdot 1} \cdot p$$
$$M_X(t) = q + pe^t$$

Therefore, the moment generating function of a Bernoulli distribution with parameter $p$ is:

$$M_X(t) = (1 - p) + pe^t$$


### Derivation of the Formula for $\mu_{1}'$ and $\mu_{2}'$

For a Bernoulli random variable $X$ with parameter $p$, the probability mass function (pmf) is given by:
$$ P(X = x) = p^x \cdot q^{1-x}, \quad x = 0, 1 $$
where $q = 1 - p$.

The $r$-th moment about the origin for a Bernoulli distribution is defined as:
$$ \mu_{r}' = E(X^r) = \sum_{x=0}^{1} x^r \cdot P(X = x) $$

#### First Moment ($\mu_{1}'$)
For $r = 1$:
$$ \mu_{1}' = E(X) = \sum_{x=0}^{1} x \cdot P(X = x) $$
$$ \mu_{1}' = 0 \cdot P(X = 0) + 1 \cdot P(X = 1) $$
$$ \mu_{1}' = 0 \cdot q + 1 \cdot p $$
$$ \mu_{1}' = p $$

#### Second Moment ($\mu_{2}'$)
For $r = 2$:
$$ \mu_{2}' = E(X^2) = \sum_{x=0}^{1} x^2 \cdot P(X = x) $$
$$ \mu_{2}' = 0^2 \cdot P(X = 0) + 1^2 \cdot P(X = 1) $$
$$ \mu_{2}' = 0 \cdot q + 1 \cdot p $$
$$ \mu_{2}' = p $$


#### Third Moment ($\mu_{3}'$)
For $r = 3$:
$$ \mu_{3}' = E(X^3) = \sum_{x=0}^{1} x^3 \cdot P(X = x) $$
$$ \mu_{3}' = 0^3 \cdot P(X = 0) + 1^3 \cdot P(X = 1) $$
$$ \mu_{3}' = 0 \cdot q + 1 \cdot p $$
$$ \mu_{3}' = p $$

#### Fourth Moment ($\mu_{4}'$)
For $r = 4$:
$$ \mu_{4}' = E(X^4) = \sum_{x=0}^{1} x^4 \cdot P(X = x) $$
$$ \mu_{4}' = 0^4 \cdot P(X = 0) + 1^4 \cdot P(X = 1) $$
$$ \mu_{4}' = 0 \cdot q + 1 \cdot p $$
$$ \mu_{4}' = p $$




```python
# Find the first four moments about origin, mean and sd of Bernoulli Distribution having probability of success 0.5
p = 0.5
x, y = bernoulli_distribution(p)

# Calculate the first four moments about the origin
moments_about_origin_0_5 = [moments_about_origin(x, y, k) for k in range(1, 5)]

# Calculate mean and standard deviation
mean_0_5 = moments_about_origin_0_5[0]
variance_0_5 = moments_about_origin_0_5[1] - mean_0_5**2
sd_0_5 = np.sqrt(variance_0_5)

print('The first four moments about origin for Bernoulli distribution with p=0.5 are:', moments_about_origin_0_5)
print('Mean:', mean_0_5)
print('Standard Deviation:', sd_0_5)
```

```python
from scipy.stats import binom
import numpy as np

# Given data
X = np.array([0, 1, 2, 3, 4])
f = np.array([28, 62, 46, 10, 4])

# Calculate the total number of trials and the probability of success
n = np.sum(f)
p_hat = np.sum(X * f) / (n * 4)

# Fit the binomial distribution
binom_dist = binom(n=4, p=p_hat)

# Calculate the expected frequencies
expected_frequencies = binom_dist.pmf(X) * n

# Print the results
print('Fitted Binomial Distribution:')
print('n:', n)
print('p:', p_hat)
print('Expected Frequencies:', expected_frequencies)

# Calculate the probability mass function for each case
pmf_values = binom_dist.pmf(X)
print('Probability Mass Function:', pmf_values)
```

```python
# Plot the probability mass function and the given data X
plt.bar(X, pmf_values, tick_label=X)
plt.xlabel('X')
plt.ylabel('Probability')
plt.title('Probability Mass Function of Fitted Binomial Distribution')
plt.show()
```
